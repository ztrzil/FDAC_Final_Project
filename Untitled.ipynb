```python
import pandas as pd
import numpy as np
import sklearn
import seaborn
import matplotlib
```

```python
senator_donation_data = pd.read_csv("./dataOutput.csv")
senator_donation_data.set_index("CID",inplace=True)
bill_votes_data = pd.read_csv("./14Cycle/billVotes.csv",na_values="?")
bill_votes_data.dropna(inplace=True)
```

```python
bills = set()
class Rep:
    def __init__(self, cid, party):
        self.cid = cid
        self.party = party
        self.votes = dict()

    def add_vote(self, num, vote):
        if vote == 1:
            self.votes[num] = 0
        elif vote == 0:
            self.votes[num] = 1
        else:
            self.votes[num] = 2
Reps = dict()

for row in bill_votes_data.values:
    cid = row[3]
    if cid not in Reps:
        Reps[cid] = Rep(cid, row[2])
    Reps[cid].add_vote(row[4], row[1])
    bills.add(row[4])
print(Reps["N00003028"].votes)
for rep in Reps:
    for bill in bills:
        if bill not in Reps[rep].votes:
            Reps[rep].votes[bill] = 2
```

```python
cleaned = open("cleaned_votes", "w")
cleaned.write("CID, Party, " + str(bills)[1:-1] + "\n")
for rep in Reps:
    r = Reps[rep]
    cleaned.write(r.cid + ", " + r.party + ", ")
    for vote in bills:
        cleaned.write(str(r.votes[vote]))
        if vote is not 705:
            cleaned.write(", ")
    cleaned.write("\n")
cleaned.close()
```

```python
#2 == Abstain, 1 == Yes, 0 == No
vote_data = pd.read_csv("cleaned_votes", index_col=False)
vote_data.set_index("CID",inplace=True)
vote_data.fillna(2,inplace=True)
data = pd.concat([vote_data,senator_donation_data])
data.fillna(0,inplace=True)
sklearn.cluster.KMeans
```

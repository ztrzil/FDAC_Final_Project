```python
import pandas as pd
import numpy as np
from collections import defaultdict


df = pd.read_csv('14Cycle/pacs14.csv')

cnt = 0
totalSum = 0
totalAvg = 0
p = defaultdict()
d = defaultdict(list)
for row in df.iterrows():
    r = row[1]
    if(r[1] > 0):
        ldon = np.log(r[1])
        d[r[0]].append(ldon)
        p[r[0]] = r[2].split()[0]
        cnt = cnt + 1
        totalSum = totalSum + r[1]
totalAvg = totalSum / cnt
```

```python
for k in d.keys():
    s = sum(d[k])
    avg = s/len(d[k])

print("There were", cnt, "donations and the average donation for the totals is $", '%.2f' % totalAvg)
tmp = 0
diff = 0
sumSqDiff = 0
for k in d.keys():
    for x in d[k]:
        x = np.exp(x)
        diff = x - totalAvg
        diffSq = diff * diff
        sumSqDiff = sumSqDiff + diffSq
variance = sumSqDiff/cnt
stdDev = variance**(1/2.0)

print("stdDev =", '%.2f' %stdDev, "and variance is", '%.2f' %variance)
```

```python
benchmark1 = -0.15
benchmark2 = -0.05
benchmark3 = 0.05
benchmark4 = 0.15
bm1 = np.exp(np.log(totalAvg)+benchmark1*np.log(stdDev))
bm2 = np.exp(np.log(totalAvg)+benchmark2*np.log(stdDev))
bm3 = np.exp(np.log(totalAvg)+benchmark3*np.log(stdDev))
bm4 = np.exp(np.log(totalAvg)+benchmark4*np.log(stdDev))

"""
print("The ranges are from 0 to ", '%.2f' %benchmark1, ",", '%.2f' % benchmark1, "to ", '%.2f' %benchmark2, ",", '%.2f' %benchmark2, '%.2f' %benchmark3, ",", '%.2f' %benchmark3,"to", '%.2f' %benchmark4, ", and over", '%.2f' %benchmark4 )
"""

f = open('dataOutput.csv', 'w') 

tmpStrg = "CID, Party, 0 to " + str('%.2f' %bm1) + ", " + str('%.2f' %bm1) + " to " + str('%.2f' %bm2) + ", " + str('%.2f' %bm2) + " to " + str('%.2f' %bm3) + ", " + str('%.2f' %bm3) + " to " +str('%.2f' %bm4) + ", over " + str('%.2f' %bm4) + '\n'
print(tmpStrg)
f.write(tmpStrg)
```

```python
for k in d.keys():
    amnts=[0 for i in range(5)]
    prcntg=[0 for i in range(5)]
    tmpSum=0
    for x in d[k]:
        y = (x - np.log(totalAvg))/np.log(stdDev)
        if y < benchmark1:
            amnts[0] = amnts[0] + x
        elif y < benchmark2:
            amnts[1] = amnts[1] + x
        elif y < benchmark3:
            amnts[2] = amnts[2] + x
        elif y < benchmark4:
            amnts[3] = amnts[3] + x
        else:
            amnts[4] = amnts[4] + x
    tmpSum = sum(amnts)
    cnt = 0
    tmpStrg = str(k) + ', ' + str(p[k])
    for i in amnts:
        prcntg[cnt] = i/tmpSum*100
        tmpStrg += ", " 
        tmpStrg += str('%.2f' %prcntg[cnt])
        cnt = cnt + 1
    tmpStrg += '\n'
    f.write(tmpStrg)
    print(tmpStrg)
f.close()
```
